## 4) Plotting a Bar chart for Car sales w.r.t Buyer's gender:

```python
car_data["Buyer Gender"].value_counts()
```

```python
plt.figure(figsize= (9,8))
car_data['Buyer Gender'].value_counts().plot(kind='bar', color = 'darkorchid',rot = 30);
plt.xlabel("Gender of the Buyer",fontsize = 12)
plt.ylabel("Count of Purchase", fontsize = 12)
plt.title(" Comparison w.r.t Gender of the Buyer", fontsize = 15)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(axis = "y",color = "silver")
plt.show()
```

#### From the Histogram we say that the Number of females who bought the car is sligtly higher than the males.

## 5)  The top 5 cars based on their sales price. :

```python
costly_cars = car_data.nlargest(5,"Sale Price","all")
costly_cars
```

#### Now, If we want to see the 'Sale Price' as the first column. (Purely for asthetic purpose)
> We can convert the data set into a list and re-arange the column names.

> Then, we finally convert the list back into a data set. Which'll have __Sale Price__ as the 1st column.

```python
car_list = costly_cars.columns.tolist()
car_list
```

```python
car_list = ['Sale Price',
 'Make',
 'Model',
 'Nickname',
 'Car Gender',
 'Buyer Gender',
 'Buyer Age',
 'Buzzword',
 'Country',
 'City',
 'Dealer Latitude',
 'Dealer Longitude',
 'Color',
 'New Car',
 'Purchase Date',
 'Discount',
 'Resell Price',
 '5-yr Depreciation',
 'Top Speed',
 '0-60 Time']
costly_cars = costly_cars[car_list]
costly_cars
```
